```python
fig,ax=plt.subplots()
frutas=['amanzana','mora azul','cereza','naranja']
conteo=[40,100,30,55]
bar_labels=['verde','azul','rojo','naranja']
bar_colors=['tab:green','tab:blue','tab:red','tab:orange']

ax.bar(frutas,conteo,label=bar_labels,color=bar_colors)

ax.set_ylabel('Suministro de fruta')
ax.set_xlabel('Nombres de frutas con colores')
ax.set_title('Suministro de frutas por tipo y color')
ax.legend(title='Color de la fruta')

plt.show()
```

Ejercicio: Escribe un programa que pregunte al usuario por los gastos que ha tenido en un rango de años y muestre por pantalla un diagrama de lineas con la evolución de sus gastos.

```python
import matplotlib.pyplot as plt
import pandas as pd

def recoger_gastos():
    gastos = {}
    print("Introduce los años de inicio y fin para los cuales deseas registrar los gastos.")
    anio_inicio = int(input("Año de inicio: "))
    anio_fin = int(input("Año de fin: "))

    for anio in range(anio_inicio, anio_fin + 1):
        gasto = float(input(f"Gasto para el año {anio}: "))
        gastos[anio] = gasto

    return gastos

gastos_usuario = recoger_gastos()

df_gastos = pd.DataFrame(list(gastos_usuario.items()), columns=['Año', 'Gasto'])

plt.figure(figsize=(10, 6))
plt.plot(df_gastos['Año'], df_gastos['Gasto'], marker='o', linestyle='-', color='g')

plt.title('Evolución de los gastos anuales')
plt.xlabel('Año')
plt.ylabel('Gasto')
plt.grid(True)

plt.show()
```
