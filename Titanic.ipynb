 El archivo titanic.csv contiene información sobre los pasajeros del Titanic. Crear un dataframe con Pandas y a partir de él generar los siguientes diagramas.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo Excel
df = pd.read_excel('titanic.xlsx')

# Ver las primeras filas del DataFrame para asegurarse de que se cargó correctamente
print(df.head())

# Asegurarse de que Seaborn y Matplotlib están listos para su uso
sns.set(style="whitegrid")  # Estilo de Seaborn para gráficos
```

*Describe al pasajero de mayor y de menor edad que viajaba en el barco

```python
# Encontrar al pasajero de mayor edad
oldest_passenger = df.loc[df['Age'].idxmax()]

# Encontrar al pasajero de menor edad
youngest_passenger = df.loc[df['Age'].idxmin()]

# Imprimir la información del pasajero de mayor edad
print("Pasajero de mayor edad:")
print(f"PassengerId: {oldest_passenger['PassengerId']}")
print(f"Survived: {oldest_passenger['Survived']}")
print(f"Pclass: {oldest_passenger['Pclass']}")
print(f"Name: {oldest_passenger['Name']}")
print(f"Sex: {oldest_passenger['Sex']}")
print(f"Age: {oldest_passenger['Age']}")
print(f"SibSp: {oldest_passenger['SibSp']}")
print(f"Parch: {oldest_passenger['Parch']}")
print(f"Ticket: {oldest_passenger['Ticket']}")
print(f"Fare: {oldest_passenger['Fare']}")
print(f"Cabin: {oldest_passenger['Cabin']}")
print(f"Embarked: {oldest_passenger['Embarked']}")

# Imprimir la información del pasajero de menor edad
print("\nPasajero de menor edad:")
print(f"PassengerId: {youngest_passenger['PassengerId']}")
print(f"Survived: {youngest_passenger['Survived']}")
print(f"Pclass: {youngest_passenger['Pclass']}")
print(f"Name: {youngest_passenger['Name']}")
print(f"Sex: {youngest_passenger['Sex']}")
print(f"Age: {youngest_passenger['Age']}")
print(f"SibSp: {youngest_passenger['SibSp']}")
print(f"Parch: {youngest_passenger['Parch']}")
print(f"Ticket: {youngest_passenger['Ticket']}")
print(f"Fare: {youngest_passenger['Fare']}")
print(f"Cabin: {youngest_passenger['Cabin']}")
print(f"Embarked: {youngest_passenger['Embarked']}")
```

*Establece cuántos pasajeros sobrevivieron y cuántos fallecieron

```python
# Contar la cantidad de sobrevivientes y no sobrevivientes
survival_counts = df['Survived'].value_counts()

# Imprimir los resultados
print(f"Cantidad de pasajeros que sobrevivieron: {survival_counts[1]}")
print(f"Cantidad de pasajeros que fallecieron: {survival_counts[0]}")
```

*Gráfica de pastel con los fallecidos y supervivientes.

```python
# Definir los colores
colors = ['red', 'green']

# Crear el gráfico de pastel
plt.figure(figsize=(4, 4))
plt.pie(survival_counts, labels=['No Sobrevivió', 'Sobrevivió'], autopct='%1.1f%%', startangle=140, colors=colors)

# Añadir un título
plt.title('Sobrevivencia de los Pasajeros del Titanic')

# Mostrar el gráfico
plt.axis('equal')  # Asegurar que el gráfico sea un círculo
plt.show()
```

*Histograma con las edades.

```python
# Eliminar valores NaN en la columna 'Age'
df_cleaned = df.dropna(subset=['Age'])

# Crear el histograma
plt.figure(figsize=(6, 4))
plt.hist(df_cleaned['Age'], bins=20, color='skyblue', edgecolor='black')

# Añadir etiquetas y título
plt.xlabel('Edad')
plt.ylabel('Cantidad de Pasajeros')
plt.title('Distribución de Edades de los Pasajeros del Titanic')

# Mostrar el histograma
plt.grid(True)
plt.show()
```

*Diagrama de barras con el número de personas en cada clase.

```python
# Contar el número de personas en cada clase y ordenar por clase
class_counts = df['Pclass'].value_counts().sort_index()

# Definir colores para cada clase
colors = ['skyblue', 'salmon', 'lightgreen']

# Crear el diagrama de barras con colores personalizados
plt.figure(figsize=(6, 4))
ax = class_counts.plot(kind='bar', color=colors)

# Añadir etiquetas y título
plt.xlabel('Clase')
plt.ylabel('Número de Pasajeros')
plt.title('Número de Pasajeros en Cada Clase del Titanic')

# Agregar el número específico a cada barra
for i, count in enumerate(class_counts):
    ax.text(i, count + 1, str(count), ha='center', color='black')

# Mostrar el diagrama de barras
plt.xticks(rotation=0)  # Rotar etiquetas del eje x
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

*Diagrama de barras con el número de personas fallecidas y supervivientes en cada clase.

```python
# Crear un DataFrame separado para los sobrevivientes y los fallecidos en cada clase
survived_class = df[df['Survived'] == 1]['Pclass'].value_counts().sort_index()
not_survived_class = df[df['Survived'] == 0]['Pclass'].value_counts().sort_index()

# Definir colores para sobrevivientes y no sobrevivientes
color_survived = 'yellow'
color_not_survived = 'purple'

# Crear el diagrama de barras
plt.figure(figsize=(6, 4))

# Barras para sobrevivientes
survived_class.plot(kind='bar', color=color_survived, position=1, width=0.4, label='Sobrevivientes', align='center')
for index, value in enumerate(survived_class):
    plt.text(index - 0.2, value + 5, str(value), color='black', fontsize=10)

# Barras para no sobrevivientes
not_survived_class.plot(kind='bar', color=color_not_survived, position=0, width=0.4, label='Fallecidos', align='center')
for index, value in enumerate(not_survived_class):
    plt.text(index + 0.2, value + 5, str(value), color='black', fontsize=10)

# Añadir etiquetas y título
plt.xlabel('Clase')
plt.ylabel('Número de Pasajeros')
plt.title('Sobrevivientes y Fallecidos por Clase en el Titanic')

# Añadir leyenda
plt.legend()

# Mostrar el diagrama de barras
plt.xticks(rotation=0)  # Rotar etiquetas del eje x
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

*Diagrama de barras con el número de personas fallecidas y supervivientes acumuladas en cada clase.


```python
# Crear un DataFrame separado para los sobrevivientes y los fallecidos en cada clase
survived_class = df[df['Survived'] == 1]['Pclass'].value_counts().sort_index()
not_survived_class = df[df['Survived'] == 0]['Pclass'].value_counts().sort_index()

# Calcular el número acumulado de sobrevivientes y no sobrevivientes por clase
survived_cumsum = survived_class.cumsum()
not_survived_cumsum = not_survived_class.cumsum()

# Crear el diagrama de barras acumuladas
plt.figure(figsize=(10, 6))

# Barras acumuladas para sobrevivientes
plt.bar(survived_cumsum.index, survived_cumsum, color='lightgreen', label='Sobrevivientes', alpha=0.7)

# Barras acumuladas para no sobrevivientes
plt.bar(not_survived_cumsum.index, not_survived_cumsum, color='salmon', label='Fallecidos', alpha=0.7, bottom=survived_cumsum)

# Añadir etiquetas y título
plt.xlabel('Clase')
plt.ylabel('Número de Personas')
plt.title('Personas Fallecidas y Sobrevivientes Acumuladas por Clase')

# Añadir leyenda
plt.legend()

# Mostrar el diagrama de barras
plt.xticks(survived_cumsum.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

*Genera un diagrama de correlación y establece si alguna de las variables presentan están conectadas.

```python
# Excluir columnas no numéricas
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
df_numeric = df[numeric_columns]

# Calcular la matriz de correlación
correlation_matrix = df_numeric.corr()

# Visualizar el diagrama de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Diagrama de Correlación de Variables')
plt.show()
```
