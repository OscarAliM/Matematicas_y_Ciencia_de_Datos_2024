#PROBLEMA 1

1. Eres el becario de un profesor muy viejito que tiene todas sus notas en una hoja de
papel, le están pidiendo que las presente como un DataFrame en Python para
poder visualizar mejor las notas de sus estudiantes y poder trabajar con ellas.
Debe haber una serie para las actividades, una para las tareas, una para quizzes,
una para exámenes y otra para evidencias. Afortunadamente su grupo es pequeño
y sólo tiene 10 alumnos. (Puedes inventar las notas y los nombres).
 Utiliza una función para mostrar los últimos 5 datos del registro.
 Usa el método columns para ver los nombres de todas las columnas.
 Usa el método dtypes para conocer los tipos de datos de nuestro DataFrame,
explica que significa el resultado que obtienes.
 Usa la función describe() para obtener la estadística descriptiva y completar la
exploración de los datos.
 Crea una función que se llame aprobados para las evidencias y otra para los
exámenes (extrae la info del DataFrame con “iloc” o “loc”) y haz que se
desplieguen las personas que tiene nota mayor o igual a 7 de manera
descendente.

```python
import pandas as pd

# Crear el DataFrame con las notas de los estudiantes
data = {
    'Estudiante': ['Juan', 'María', 'Pedro', 'Ana', 'Luis', 'Sara', 'Pablo', 'Lucía', 'Carlos', 'Laura'],
    'Actividades': [8.5, 9.0, 7.5, 6.0, 8.0, 9.5, 7.0, 6.5, 8.0, 9.0],
    'Tareas': [7.0, 8.0, 6.5, 7.5, 8.5, 9.0, 7.5, 6.0, 8.0, 7.0],
    'Quizzes': [6.5, 7.0, 8.0, 7.5, 6.0, 8.0, 9.0, 7.5, 8.5, 6.5],
    'Exámenes': [7.5, 8.0, 9.0, 6.0, 7.0, 8.5, 9.5, 7.0, 8.0, 9.0],
    'Evidencias': [9.0, 7.5, 8.0, 9.5, 7.0, 8.0, 7.5, 8.5, 9.0, 7.5]
}

df = pd.DataFrame(data)

# Mostrar los últimos 5 datos del registro
def mostrar_ultimos_datos(dataframe):
    return dataframe.tail(5)

ultimos_datos = mostrar_ultimos_datos(df)
print(ultimos_datos)

# Ver los nombres de todas las columnas
columnas = df.columns
print(columnas)

# Conocer los tipos de datos de nuestro DataFrame
tipos_de_datos = df.dtypes
print(tipos_de_datos)

# Obtener la estadística descriptiva de los datos
estadisticas_descriptivas = df.describe()
print(estadisticas_descriptivas)

# Función para mostrar aprobados en las evidencias
def aprobados_evidencias(dataframe):
    aprobados = dataframe[dataframe['Evidencias'] >= 7].sort_values('Evidencias', ascending=False)
    return aprobados

aprobados_en_evidencias = aprobados_evidencias(df)
print("Aprobados en evidencias:")
print(aprobados_en_evidencias)

# Función para mostrar aprobados en los exámenes
def aprobados_examenes(dataframe):
    aprobados = dataframe[dataframe['Exámenes'] >= 7].sort_values('Exámenes', ascending=False)
    return aprobados

aprobados_en_examenes = aprobados_examenes(df)
print("Aprobados en exámenes:")
print(aprobados_en_examenes)
```

 Usa el método shape para averiguar cuantas filas y columnas tiene nuestro

 Usa la función info() que nos regresa la información completa de los datos

 Estable en promedio cuantas calorías tienen los alimentos de la cadena y
también da el valor máximo de calorías y azúcares que aportan sus
alimentos.
 Muestra en la pantalla el alimento con identificador 173
 Haz que se muestren en pantalla sólo los 'Smoothies & Shakes' y
ordénalos alfabéticamente.


```python
menu = pd.read_excel('menu_1.xlsx')

# Función para probar que el archivo se cargó correctamente
def probar_carga(dataframe):
    return dataframe.head(5)

primeros_registros = probar_carga(menu)
print(primeros_registros)

# Obtener el número de filas y columnas del menú
num_filas, num_columnas = menu.shape
print("Número de filas:", num_filas)
print("Número de columnas:", num_columnas)

# Obtener información completa del DataFrame
informacion = menu.info()
print(informacion)

# Calcular el promedio, valor máximo de calorías y azúcares
promedio_calorias = menu['Calories'].mean()
max_calorias = menu['Calories'].max()
max_azucar = menu['Sugars'].max()
print("Promedio de calorías:", promedio_calorias)
print("Valor máximo de calorías:", max_calorias)
print("Valor máximo de azúcares:", max_azucar)

# Mostrar el alimento con identificador 173
# Obtener el nombre del artículo en la fila 173
nombre_articulo_173 = menu.loc[172, 'Item']
print("Nombre del artículo en la fila 173:", nombre_articulo_173)

# Filtrar y ordenar los 'Smoothies & Shakes' alfabéticamente
smoothies_shakes = menu[menu['Category'] == 'Smoothies & Shakes'].sort_values('Item')
print("Smoothies & Shakes:")
print(smoothies_shakes)
```
